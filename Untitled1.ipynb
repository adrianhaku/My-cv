```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ---- First system (r, phi) ----
def system_1(t, y, omega):
    r, phi = y  # y[0] = r, y[1] = phi
    dr_dt = omega**2 * r
    dphi_dt = 0  # Phi does not change with time
    return [dr_dt, dphi_dt]

# Given conditions
r0 = 4  # Initial distance in cm
phi0 = 0  # Initial angle
T = 10 * 60  # 10 minutes in seconds
omega = (6 * np.pi) / T  # 3 full rotations in 10 min

tspan_1 = np.linspace(0, 600, 300)  # First 600 seconds

# Solve numerically
sol_1 = solve_ivp(system_1, [0, 600], [r0, phi0], args=(omega,), t_eval=tspan_1)

# Extract solutions
t_values_1 = sol_1.t
r_values = sol_1.y[0]

# Convert to Cartesian coordinates (x, y)
x_values = r_values * np.cos(omega * t_values_1)
y_values = r_values * np.sin(omega * t_values_1)

# ---- Second equation (x'') ----
def system_2(t, y):
    x, dx = y  # y[0] = x, y[1] = dx/dt
    ddx = -x + 2*dx  # Second derivative equation rearranged
    return [dx, ddx]

# Initial conditions for x equation
t0_2 = 2
x0 = [1, -2]  # x(2) = 1, x'(2) = -2
tspan_2 = np.linspace(t0_2, 3, 100)

# Solve numerically
sol_2 = solve_ivp(system_2, [t0_2, 3], x0, t_eval=tspan_2, method='RK45')

# Extract solutions
t_values_2 = sol_2.t
x_values_2 = sol_2.y[0]

# ---- Plotting Results ----
plt.figure(figsize=(12, 5))

# First system plot (particle movement)
plt.subplot(1, 2, 1)
plt.plot(x_values, y_values, 'b-', linewidth=2)
plt.xlabel("X position (cm)")
plt.ylabel("Y position (cm)")
plt.title("Motion of Particle in Rotating Tube")
plt.grid(True)

# Second system plot (x'' - 2x' + x = 0)
plt.subplot(1, 2, 2)
plt.plot(t_values_2, x_values_2, 'r-', linewidth=2, label="x(t)")
plt.xlabel("Time (t)")
plt.ylabel("Solution x(t)")
plt.title("Numerical Solution of x'' - 2x' + x = 0")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
